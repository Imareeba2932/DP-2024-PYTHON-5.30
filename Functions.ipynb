```python
#non parameterized function
def my_func(): #Function definition
    print("Hello from my_func")
```

```python
my_func() #Function callable
my_func()
my_func()
```

```python
#parameterized function
#non return function
def sum(num1, num2): #parameters
    print(num1 + num2)
```

```python
a = sum(10, 20) #arguments
```

```python
print(a)
```

```python
def name(fname):
    print("Hello", fname)
```

```python
name("Areeba")
```

```python
#return function
def func(x):
    return 5*x

a = func(3)
print(a)
```

```python
print(a)
```

```python
#default parameterized function
def my_country(country = "India"):
    print("I am from "+ country)

my_country("USA")
my_country("Pakistan")
my_country("UK")
my_country()
```

```python
print(math.sqrt(16))
print(math.pow(2, 3))
print(math.pi)
print(math.e)
print(math.factorial(5))
print(math.gcd(12, 15))
print(math.sin(90))
print(math.cos(90))
print(math.tan(90))
print(math.lcm(12,15))
print(math.log(10))
```

```python
from random import random, randint, choice, shuffle
```

```python
print('get a random number between 0 and 1:', random())
print('get a random number between 1 and 100:', randint(1, 100))
print('get a random number from a list:', choice([1, 2, 3, 4, 5]))
names = ['Areeba', 'Ali', 'Ahmed', 'Asad', 'Ahsan']
shuffle(names)
print('shuffle a list:', names)
```

```python
from datetime import datetime as dt
```

```python
print('Current Date and Time :', dt.now())
print('Current Date :', dt.now().date())
print('Current Time :', dt.now().time())
print('Current Year :', dt.now().year)
print('Current Month :', dt.now().month)
print('Current Day :', dt.now().day)
print('Current Hour :', dt.now().hour)
print('Current Minute :', dt.now().minute)
print('Current Second :', dt.now().second)
print('Current Microsecond :', dt.now().microsecond)
```

```python
date = dt.strftime(dt.now(), '%d/%m/%Y')
print('Date in dd-mm-yyyy format :', date)
```

```python
time = dt.strftime(dt.now(), '%H:%M:%S')
print('Time in 24 hour format :', time)
```
